- All these features has outliers so we replace missing values with median

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df['income'].fillna(df['income'].median(), inplace=True)
```

```python
df['savings'].fillna(df['savings'].median(), inplace=True)
```

```python
df['monthly_expenses'].fillna(df['monthly_expenses'].median(), inplace=True)
```

```python
df['credit_score'].fillna(df['credit_score'].median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df['education'].value_counts()
```

- We can see that there is inconsistent spelling (Bachlors).
- But model treat bachelors and bachlors both as seperate categories
- So, we will replace it with correct spelling

```python
df['education'] = df['education'].replace({
    'Bachlors' : 'Bachelors'
})
```

```python
df['education'].value_counts()
```

#### Encode categorical variables (one-hot encoding / ordinal encoding).

```python
df.info()
```

- We use One-Hot Encoding for marital_status and Home_ownership columns . since, these columns do not have hierarchy
- We use ordinal encoding for Education column. since, it has hierarchy
- Drop customer_id. As it donot have relation with the target_default_risk
- Change sigup_date into datetime format represents the customer days since signedup

```python


marital_status = pd.get_dummies(df.marital_status,dtype = 'int', drop_first=True)
df = pd.concat([df,marital_status], axis=1)
```

```python
home_ownership = pd.get_dummies(df.home_ownership,prefix = 'H',dtype='int',drop_first=True)
df = pd.concat([df,home_ownership],axis=1)
```

```python
region = pd.get_dummies(df.region,dtype='int',drop_first=True)
df = pd.concat([df,region],axis=1)
```

```python
df.info()
```

```python
dict1 = {
    'HS'        : 1,
    'Bachelors' : 2,
    'Masters'   : 3,
    'PhD'       : 4,
    'Other'     : 5
}

df['education'] = df['education'].map(dict1)
```

```python
df.info()
```

```python
df['signup_date'] = pd.to_datetime(df['signup_date'])
df['days_since_signup'] = (pd.Timestamp('today') - df['signup_date']).dt.days
df.drop(columns=['signup_date'], inplace=True)
```

```python
df.drop(columns = ['customer_id','home_ownership','marital_status','region'],inplace=True)
```

```python
df.info()
```

#### Scaling Numeric features

```python
# Scaling numeric features 

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Get numeric columns
num_cols = df.select_dtypes(include='number').columns

# Remove target column from scaling list
num_cols = num_cols.drop('target_default_risk')

# Scale only numeric features
df[num_cols] = scaler.fit_transform(df[num_cols])
```

```python
df.info()
```

```python
df['loan_to_income_ratio'] = df['loan_amount'] / df['income']   # feature transforming will enhance predictive power
```

#### Outlier treatment Winsorization

```python
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3-Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR


    df[col] = np.where(df[col]< lower, lower, df[col])
    df[col] = np.where(df[col]> upper, upper,df[col])

    
```

# MODEL BUILDING

```python
# TRAIN TEST SPLIT

from sklearn.model_selection import train_test_split
x = df.drop(columns = ['target_default_risk'])
y = df['target_default_risk']
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
```

### 1.LOGISTIC REGRESSION

```python
x_train.isnull().sum()
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=250)
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
y_pred_lg = model.predict(x_test)
```

```python
print(classification_report(y_pred_lg,y_test))
acc_lg = accuracy_score(y_test, y_pred_lg)
```

### 2. DECISION TREE

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(random_state=42)
model.fit(x_train,y_train)
```

```python
y_pred_dec = model.predict(x_test)
print(classification_report(y_pred_dec,y_test))
acc_dec = accuracy_score(y_pred_dec, y_test)
```

### 3. SUPPORT VECTOR MACHINE

```python
from sklearn.svm import SVC
model = SVC(C= 4, gamma=0.01)
model.fit(x_train,y_train)
```

```python
y_pred_svm = model.predict(x_test)
print(classification_report(y_pred_svm,y_test))
acc_svm = accuracy_score(y_pred_svm,y_test)
```

### 4.RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier
model_rf = RandomForestClassifier(n_estimators=700,random_state=42,  min_samples_leaf=3)
model_rf.fit(x_train,y_train)
```

```python
y_pred_rf = model_rf.predict(x_test)
print(classification_report(y_pred_rf,y_test))
acc_rf = accuracy_score(y_pred_rf,y_test)
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE()
x_train_smote, y_train_smote = smote.fit_resample(x_train,y_train)
```

```python
y_train_smote.value_counts()
```

```python
model_rf.fit(x_train_smote,y_train_smote)
```

```python
y_pred_rf_smote = model_rf.predict(x_test)
print(classification_report(y_pred_rf_smote, y_test))
```

### 5.XGBOOST

```python
from xgboost import XGBClassifier
model_xgb = XGBClassifier()
model_xgb.fit(x_train_smote,y_train_smote)
```

```python
y_pred_xg= model_xgb.predict(x_test)
print(classification_report(y_pred_xg,y_test))
acc_xg = accuracy_score(y_pred_xg, y_test)
```

## COMPARING ALL MODELS ACCURACY

```python
results = {
    "Logistic Regression": acc_lg,
    "Decision Tree": acc_dec,
    "SVM": acc_svm,
    "Random Forest": acc_rf,
    "XGBoost": acc_xg
}

pd.DataFrame(results, index=["Accuracy"]).T.sort_values(by="Accuracy", ascending=False)
```

### Hyper parameter tuning

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score

# ---------------------------
# RANDOM FOREST (GridSearch)
# ---------------------------

param_grid_rf = {
    'n_estimators': [300, 500],
    'max_depth': [None, 10, 20],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2']
}

grid_rf = GridSearchCV(
    RandomForestClassifier(random_state=42),
    param_grid_rf,
    cv=5,
    scoring='accuracy',
    n_jobs=-1
)

grid_rf.fit(x_train, y_train)

best_rf = grid_rf.best_estimator_

print("Best Random Forest Params:", grid_rf.best_params_)
print("Random Forest Accuracy:",
      accuracy_score(y_test, best_rf.predict(x_test)))

```

```python
acc_rf_best = accuracy_score(y_test, best_rf.predict(x_test))
```

```python
print(acc_rf_best)
```

```python
# XGBOOST (RandomizedSearch)
# ---------------------------

param_dist_xgb = {
    'n_estimators': [300, 500, 700],
    'learning_rate': [0.01, 0.05, 0.1],
    'max_depth': [4, 6, 8],
    'subsample': [0.8, 1],
    'colsample_bytree': [0.8, 1],
    'gamma': [0, 0.1, 0.3]
}

random_xgb = RandomizedSearchCV(
    XGBClassifier(random_state=42, eval_metric='logloss'),
    param_distributions=param_dist_xgb,
    n_iter=20,
    cv=5,
    scoring='accuracy',
    random_state=42,
    n_jobs=-1
)

random_xgb.fit(x_train, y_train)

best_xgb = random_xgb.best_estimator_

print("\nBest XGBoost Params:", random_xgb.best_params_)
print("XGBoost Accuracy:",
      accuracy_score(y_test, best_xgb.predict(x_test)))
```

```python
acc_xgb_best = accuracy_score(y_test, best_xgb.predict(x_test))
print(acc_xgb_best)
```
